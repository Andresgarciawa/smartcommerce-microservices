# Notebook para leer y añadir más datos al archivo Excel `Libro1-pruebas.xlsx`
Este notebook permite inspeccionar el archivo existente, agregar filas nuevas y guardar los cambios.

```python
import pandas as pd
from pathlib import Path

excel_path = Path('Libro1-pruebas.xlsx')
print('Excel path:', excel_path.resolve())

df = pd.read_excel(excel_path)
df.head()
```

```python
print('Columnas:', list(df.columns))
print('Filas actuales:', len(df))

df
```

```python
new_rows = [
    {
        'external_code': '3',
        'book_reference': 'aventura',
        'quantity_available': 15,
        'quantity_reserved': 1,
        'condition': 'nuevo',
        'defects': '',
        'observations': 'llegada nueva',
    },
    {
        'external_code': '4',
        'book_reference': 'fantasia',
        'quantity_available': 12,
        'quantity_reserved': 0,
        'condition': 'nuevo',
        'defects': '',
        'observations': 'stock inicial',
    },
    {
        'external_code': '5',
        'book_reference': 'misterio',
        'quantity_available': 5,
        'quantity_reserved': 0,
        'condition': 'reacondicionado',
        'defects': 'leves',
        'observations': 'revisado',
    },
]

new_df = pd.concat([df, pd.DataFrame(new_rows)], ignore_index=True)
new_df.tail()
```

```python
new_df.to_excel(excel_path, index=False)
print('Archivo actualizado con nuevas filas. Total de filas:', len(new_df))
```
